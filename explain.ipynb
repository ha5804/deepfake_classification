# 1.파이프 라인

입력 -> 얼굴 검출 -> 랜드마크 기반 정렬 -> 얼굴 crop -> vit로 프레임별 추론 -> 평균 -> 제출

# 2.환경 및 라이브러리 import

(1) path: os독립적인 파일 경로 관리

(2) typing: 코드 가독성 + 유지 보수용

(3) cv2 : 이미지/ 비디오 로딩 & warp

(4) dlib : 얼굴 검출 + 랜드마크

(5) torch : 모델 추론

(6) F.softmax : 확률 계산

(7) PIL.Image : huggingface processor 입력 형식

(8) tqdm : 전처리 / 추론 진행 상황 확인

(9) vitforimageclassification : 사전 학습된 딥 페이크 vit

(10) similarityTransorm : 얼굴 정렬 핵심
